# Matplotlib Summary

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

## Plot

```python
x = np.arange(1, 6)
y = np.arange(2, 11, 2)
plt.plot(x, y, "red")
plt.show()
```

## Subplot

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, "red")

plt.subplot(2, 2, 2)
plt.plot(y, x, "blue")

plt.subplot(2, 2, 3)
plt.plot(x, y**2, "green")

plt.subplot(2, 2, 4)
plt.plot(x, y**3, "orange")

plt.show()
```

## Figures & Axes

```python
a = np.arange(1,6)
b = np.arange(2, 11, 2)
```

```python
#### Create a figure
fig = plt.figure(figsize=(10, 10))

#### Define axes

####                ([left_padding, bottom_padding, width, height])
axes1 = fig.add_axes([0.3, 0.5, 0.6, 0.6])
axes2 = fig.add_axes([0.4, 0.7, 0.3, 0.3])

axes1.plot(a, b**4)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(b, a, "red")
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")
plt.show()
#### A nested shape is formed, as shown below.
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
#### Axes are in the array
axes
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

#### Axes are in the array
for ax in axes:
    ax.plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(a, b)
axes[0].set_title("First Axes")
axes[1].plot(a, b**3)
axes[1].set_title("Second Axes")

#### Graphs padding 
plt.tight_layout(pad=5)
```

```python
#### Setting figure size
fig = plt.figure(figsize=(10,4))

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b**5, color="purple")
```

```python
#### Setting figure size
fig = plt.figure(figsize=(2,4))

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b**5, color="purple")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 8))

axes[0].plot(a, b**4, "red")
axes[0].set_title("First Graph")
axes[1].plot(a, b**0.5, "yellow")
axes[1].set_title("Second Graph")

plt.tight_layout(pad=3)
```

## Adding Legend

```python
fig = plt.figure(figsize=(10,4))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b**0.5, color="red", label="Square Root")
axes.plot(a, b**2, color="yellow", label="Square")
axes.plot(a, b**3, color="green", label="Cube")

plt.legend()
```

## Customizing the Figures

```python
fig = plt.figure(figsize=(8, 6))
axes = fig.add_axes([0, 0, 1, 1])

#### Edit line width...
#### <<<<< linewidth = lw >>>>>

axes.plot(a, b**2, color="red", lw=10)
axes.plot(a, b**2, color="red", linewidth=10)
```

```python
fig = plt.figure(figsize=(8, 6))
axes = fig.add_axes([0, 0, 1, 1])

#### Edit line style...
#### <<<<< linestyle = ls >>>>>

axes.plot(a, b**3, color="green", lw="5", linestyle="--")
axes.plot(a, b**2, color="black", lw="5", linestyle=":")
axes.plot(a, b**2.5, color="orange", lw="5", linestyle="-.")
```

```python
fig = plt.figure(figsize=(8, 6))
axes = fig.add_axes([0, 0, 1, 1])

#### Adding marker...

axes.plot(a, b**3, color="red", lw="1", linestyle="--", marker="o", 
          markersize=20, markerfacecolor="green", markeredgecolor="yellow", markeredgewidth=10)
```

```python
fig = plt.figure(figsize=(8, 6))
axes = fig.add_axes([0, 0, 1, 1])

#### Adding limit on axes

axes.plot(a, b**3, color="red", lw="1", linestyle="--", marker="o", 
          markersize=20, markerfacecolor="green", markeredgecolor="yellow", markeredgewidth=3)
axes.set_xlim(0, 20)
axes.set_ylim(0, 2000)
```

## Examples

```python
x = np.arange(0, 50)

fig = plt.figure(figsize=(8,8))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x*2, color="red")
```

```python
fig = plt.figure(figsize=(8, 6))
axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])
axes1.plot(x, x*2)
axes2.plot(x, x/2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
axes[0].plot(x, x*2, color="red", lw=5, marker="o", markersize="10", markerfacecolor="black", markeredgecolor="red")
axes[1].plot(x, x**3, color="purple", ls="-.")
plt.tight_layout(pad=2)
```
